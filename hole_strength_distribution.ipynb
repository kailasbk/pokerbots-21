```python
from cards import all_cards_excluding, draw_random_cards, sort_cards, hole_strengths
import matplotlib.pyplot as plt
import numpy as np
```

```python
deck = all_cards_excluding()
ITERS = 500000
sample_hole_strengths = []
# sample_hole_strengths = np.empty(ITERS)
for _ in range(ITERS):
    hole_cards = draw_random_cards(deck, 2)
    sorted = sort_cards(hole_cards)

    if hole_cards[0][1] == hole_cards[1][1]:
        prob = float(hole_strengths[str(sorted[1][0] + sorted[0][0] + 's')])
    else:
        prob = float(hole_strengths[str(sorted[1][0] + sorted[0][0] + 'o')])

    sample_hole_strengths.append(prob)
    # np.append(sample_hole_strengths, prob)
```

```python
plt.hist(sample_hole_strengths, bins=10)
```

```python
plt.hist(sample_hole_strengths, bins=30, range=(0.3, 1), density=True)
```

```python
def board_hole_strength(board):
    line_split = [line.split(' ') for line in open('board' + str(board))]
    card_array_str = []
    for line in line_split:
        middle_two = line[2:4]
        cards = [card.strip('[]') for card in middle_two]
        card_array_str.append(cards)

    shown_holes_b = []
    for sorted in card_array_str:
        if hole_cards[0][1] == hole_cards[1][1]:
            prob = float(hole_strengths[str(sorted[1][0] + sorted[0][0] + 's')])
        else:
            prob = float(hole_strengths[str(sorted[1][0] + sorted[0][0] + 'o')])
        
        shown_holes_b.append(prob)
        
    return shown_holes_b
```

```python
plt.xlabel('Hole Strength')
plt.ylabel('Probability Density')
plt.hist(board_hole_strength(1), bins=30, range=(0.3, 1), density=True)
plt.title('Hands Shown on Board 1')
```
